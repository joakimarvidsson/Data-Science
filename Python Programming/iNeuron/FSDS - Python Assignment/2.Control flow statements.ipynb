**Q1. Find the largest number from the given list, use for loop.**

**numbers = [12, 75, 150, 180, 145, 525, 50]**

```python
numbers = [12, 75, 150, 180, 145, 525, 50]

large = numbers[0]

for i in numbers:
  if i > large:
     large = i
print("Greatest number: ", large)
```

**Q3. Calculate the sum of all numbers from 1 to a given number**

Expected Output:

Enter number 10

Sum is:  55

```python
#define starts from 1

start = 1
end = int(input("Enter the ending number: "))
sum = 0 # initialize result
for num in range(start,end+1,1):
  sum=sum+num
print("Sum of 1 to", end, "numbers is: ", sum)
```

**Q4. Write a program to display only those numbers from a list that satisfy the following conditions**
- The number must be divisible by five

- If the number is greater than 150, then skip it and move to the next number

- If the number is greater than 500, then stop the loop

```python
# Given:

numbers = [12, 75, 150, 180, 145, 525, 50]
```

```python
a = [12, 75, 150, 180, 145, 525, 50]
b = []
for i in a: 
    #If the number is greater than 150, then skip it and move to the next number
    if i > 150:
        if i > 500: #If the number is greater than 500, then stop the loop
            break
        continue
    if i % 5 == 0: #The number must be divisible by five
        b.append(i)
        
print(b)
```
