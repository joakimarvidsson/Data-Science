#### **2. write a function which  will be able to print an index of list element without using an index function**


```python
def index_element(list):
    for i in range(len(list)):
        print(f"{list[i]} is index= {i}")
index_element(['a','b','c','d','e'])
```

#### **3. write a function which will be able to print an ip address of your system**


```python
#Import the socket library
import socket

#Then print the value of the IP into the print() function your IP address.
def ip():
    hostname = socket.gethostname()
    ipaddrs = socket.gethostbyname(hostname)

    print("Your Computer Name is:" + hostname)
    print("Your Computer IP Address is:" + ipaddrs)
ip()
```

#### **5. write a function which will take input as a list with any kind of numeric value and give an out as a multiplication of**
#### **all the numeric data l = [3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']**


```python
def numeric_multiplication(lst):
    mul=1
    for i in lst:
        if type(i)!=str:
            mul*=i
    return mul
print("numeric multiplication of [3.5, 6.56, 4,5,'sudh' , 'ineuron' , 'fsda bootcamp 2.0'] = ",numeric_multiplication([3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']))
print("numeric multiplication of [1,2,3,4,5,6] = ",numeric_multiplication([1,2,3,4,5,6]))
```

```python
import imaplib
import email
from imap_tools import MailBox, AND
def read_mail():
    server="imap.gmail.com"
    user='****'
    password='******'

    mb = MailBox(server).login(user, password)
    messages = mb.fetch(criteria=AND(seen=False),
                            mark_seen=False,
                            bulk=True)
   
    for msg in messages:
        # Print from and subject
        print(msg.from_, ': ',msg.subject)
        # Print the plain text (if there is one)
        print(msg.text)
        
        break
read_mail()
```

#### **7 . write a function which will be able to send a mail to anyone**


```python
import smtplib
from email.message import EmailMessage
def send_message():

    user='***'
    password='****'

    # create email
    msg = EmailMessage()
    msg['Subject'] = "Testing for ineuron Assignment"
    msg['From'] = user
    msg['To'] = "nirmalvignu@gmail.com"
    msg.set_content('''This email is in part of ineuron assignment to send email using python functions.
    Have a great day
                    ''')

    # send email
    smtp=smtplib.SMTP_SSL('smtp.gmail.com', 465)
   
    smtp.login(user, password)
    smtp.send_message(msg)
    smtp.quit()
send_message()
```
